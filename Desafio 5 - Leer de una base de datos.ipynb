```python
import requests
import json
import pandas as pd

def load_json_as_dataframe(url):
    data_dict = json.loads(requests.get(url).text ) #load json from url as dict
    df = pd.json_normalize(data_dict) #save json dict as a normalized df
    return df

df = load_json_as_dataframe('https://jsonplaceholder.typicode.com/users')
```
